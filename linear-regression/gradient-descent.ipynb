## Gradient Descent Optimization
Gradient descent is an optimization technique to find the maxima/minima for a function. For linear regression, we use it to find the minima for the cost function.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(X):
    return (X-10)**2 + 2*X - 5
```

```python
X = np.arange(20)
Y = function(X)
print(f'x - {X}')
print(f'f(x) - {Y}')
```

```python
# plotting the function
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

In gradient descent, we start with any arbitrary value and then iteratively move the value of the parameter to reach closer to the minima. Here as you can see, the minima is somewhere around 8.5-9. After assuming an initial value it computes the slope of the curve and based on the slope it moves x to right or left. Here if we start with x = 0, the slope of the curve at x = 0 is negative, so we need to increment x to reach closer to the minima. The slope is computed by taking the derivative of the function.

For each iteration,
x = x - n*f'(x)

n - learning rate, controls how fast/slow the value of x changes
f'(x) - derivative of y wrt x, to get the slope of the function

```python
# running multiple iterations
plt.plot(X,Y)

x = 0 
y = function(x)
learn_rate = 0.2
plt.scatter(x,y)

for i in range(20):
    x = x - learn_rate*(2*x-18)
    y = function(x)
    print(f'After iteration {i+1}, (x = {x}, y = {y})')
    plt.scatter(x,y)

plt.show()
```

## Importance of learning rate
Learning rate controls how fast or slow the value of the x gets updated with each iteration, if it too low, we might need large number of iterations to get to the minima, if it is too high, we might diverge rather than converging to minima. In the example shown below, if we take learning rate 1.1, we diverge rather than converging.

```python
# running multiple iterations
plt.plot(X,Y)

x = 0 
y = function(x)
learn_rate = 1.1
plt.scatter(x,y)

for i in range(5):
    x = x - learn_rate*(2*x-18)
    y = function(x)
    print(f'After iteration {i+1}, (x = {x}, y = {y})')
    plt.scatter(x,y)

plt.show()
```
